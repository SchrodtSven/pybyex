```python
import scipy.stats as st 
from scipy.stats import binom, norm
import numpy as np
import matplotlib.pyplot as plt



import pandas as pd

from math import sqrt
```

```python
p_value = binom.cdf(30, 1000, 0.10) #"die WK dafür, dass: höchstens 30 leute spielen wow"

# Parameter
n = 1000
p0 = .1
alpha = 0.5

# Erwartungswert und Standardabweichung unter H0 
mu = n * p0 # Da wir davon ausgehen, dass Binomialverteilt! also E(X) = n * p 
sigma = np.sqrt((n * p0)*(1 - p0)) # Var(X) = n * p * q -> sqrt(Var(X)) = sigma 

# Kritischer z-Wert und kritischer Schwellenwert 
z_alpha = norm.ppf(alpha) # wichtig umkehrfunktion, ALSO gibt zu einer WK% den representativen Wert der zugehörigen Zufallsgröße 
k = mu + z_alpha * sigma
k_cc = np.floor(k + 0.5) # auf ganze erfolge Runden

# x achse 
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000) # x achse erstellt und an unsere verteilungsfunk angepasst
y = norm.pdf(x, mu, sigma) # N(mu, sigma)

# plot 
with plt.xkcd():
    fig, ax = plt.subplots()
    ax.plot(x,y, label="Normalapproximation von X")

    # Kritsche Region (linker alpha Bereich schraffieren)

    x_fill = np.linspace(mu - 4 * sigma, k_cc, 300)
    ax.fill_between(x_fill, norm.pdf(x_fill, mu, sigma), alpha=.4, label=f"{alpha:.2f}")

    # Kritischen Werte markieren 
    ax.axvline(k_cc, linestyle="--", label = f"kritischer Wert k = {int(k_cc)}")

    # Achsenbeschriftung und Titel 
    ax.set_xlabel("Anzahl Spieler X")
    ax.set_ylabel("Dichte")
    ax.set_title("Einseitiger Test: p < 10% (n = 1000) Normalapproximation")
    ax.legend() 

plt.show()


print(f"mu = {mu:.2f}, sigma = {sigma:.2f}")
print(f"kritischer wert k : {int(k_cc)}")


```
