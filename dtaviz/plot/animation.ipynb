```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation as anima
plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['figure.dpi'] = 150  
plt.ioff()
fig, ax = plt.subplots()               # leere Standardgrafik
ax.set_xlim([0, 2*np.pi])
ax.set_ylim([-1.1, 1.1])
x = np.linspace(0, 2 * np.pi, 100)     # erstellt selbständig num-Werte aus dem Interval aus start/stop/num, 
                                       # muss mir nix überlegen
y = np.sin(x)                          # berechnet für alle x-Werte den dazugehörigen y-Wert entsprechend dem sinus-x

#plt.plot(x, y);              # fig (erstellten leeren Graph) mit Werten gefüllt
                             # ohne ; erscheint diese Ausgabe: [<matplotlib.lines.Line2D at 0x1e42ce17110>]
                             # sie beschreibt den letzten Wert/Datentyp der ausgeführt wird.
                             # Dies Information wird ausgeblendet, wenn dem plt.plot() ein ; angehängt 
                             # wird oder plt.show() ausgeführt wird

sinegraph, = ax.plot([], [])

def init():
    sinegraph.set_data([], [])
    return sinegraph

def upd(i):
    sinegraph.set_data(x[:i], y[:i])    # die x und y Werte werden bei jedem Aufruf um 1 erhöht
    ax.plot(x[i], y[i])             # die x und y Werte werden bei jedem Aufruf annimiert

ani = anima(fig, upd, frames=len(x), interval=50, init_func=init)
from IPython.display import HTML
HTML(ani.to_jshtml())
plt.show()
```
